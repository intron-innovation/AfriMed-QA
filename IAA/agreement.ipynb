```python

# Find the duplicated questions
# create a table of the response of both respondet to each question
# Create a column renaming the response for each question as Rater 1 and Rater 2

dups = []

for i in range(len(df2)):
    q1 = df2['question'][i]
    if df2[df2['question'] == q1].shape[0] == 2:
        if q1 in dups:
            pass
        else:
            dups.append(q1)

quest = pd.DataFrame()
for i in dups:
    if df2[df2['question'] == i].shape[0] == 2:
        n = df2[df2['question'] == i].reset_index(drop=True)

        quest = pd.concat([quest,n],axis=0)

quest.reset_index(inplace=True)
quest.rename(columns={"index":"rater"},inplace=True)
quest['rater'] = quest['rater'].replace({0:"Rater 1", 1:"Rater 2"})
```

```python
# Pivot the dataset for question evaluation Score
correct_pivot = quest.pivot(index="question", columns="rater", values="Correct and consistent with scientific consensus")
african_pivot = quest.pivot(index="question", columns="rater", values="Requires African local expertise")
omission_pivot = quest.pivot(index="question", columns="rater", values="Omission of relevant info")
relevant_pivot = quest.pivot(index="question", columns="rater", values="Includes irrelevant- wrong- or extraneous information")
logic_pivot = quest.pivot(index="question", columns="rater", values="Evidence of correct reasoning or logic")
bias_pivot = quest.pivot(index="question", columns="rater", values="Indication of demographic bias")
harm_pivot = quest.pivot(index="question", columns="rater", values="Possibility of harm")
quality_pivot = quest.pivot(index="question", columns="rater", values="Question Quality")
format_pivot = quest.pivot(index="question", columns="rater", values="Formatting- style- structure or grammar issues")

```

```python
#concatenate the response for all the question evaluation

iaa_full= pd.concat([bias_pivot,harm_pivot,logic_pivot,format_pivot,african_pivot,\
                     correct_pivot,quality_pivot, omission_pivot,relevant_pivot])
```

```python
#Using the Fleiss Kappa inter rater

from statsmodels.stats.inter_rater import fleiss_kappa

iaa_full_matrix = iaa_full.apply(pd.Series.value_counts, axis=1).fillna(0).astype(int)

# Calculate Fleiss' Kappa
kappa = fleiss_kappa(iaa_full_matrix.values)
print("Overall Fleiss' Kappa:", kappa)
```

```python
# Kappa Inter rater agreement for individual metrics

models = [bias_pivot,harm_pivot,logic_pivot,format_pivot,african_pivot,correct_pivot,quality_pivot, omission_pivot,relevant_pivot]
m = ['bias','harm','logic','format','african','correct','quality', 'omission','relevant']
def kappa(data,criteria):
    data_mat = data.apply(pd.Series.value_counts, axis=1).fillna(0).astype(int)
    kappa = fleiss_kappa(data_mat.values)
    print(f"{criteria} Kappa IAA:", kappa)
```

```python
for i in range(len(m)):
    kappa(models[i],m[i])
```

```python
# Using the Krippendorff metric

import krippendorff

# Convert the pivot table to a list of lists for alpha calculation
clarity_scores = iaa_full.values.T  # Transpose for Krippendorff's alpha

# Calculate Krippendorffs Alpha
alpha = krippendorff.alpha(reliability_data=clarity_scores, level_of_measurement='interval')
print("Clarity Krippendorff's Alpha:", alpha)
```

```python
# krippendorff for the individual evaluation metric
 
def krip(data,criteria):
    data_scores = data.values.T
    alpha = krippendorff.alpha(reliability_data=data_scores, level_of_measurement='interval')
    print(f"{criteria} Krippendorff's Alpha:", alpha)
```

```python
for i in range(len(models)):
    krip(models[i],m[i])
```

```python
# Implementing Krippendorf from ascratch

def krippendorffs_alpha(data):
    """
    Compute Krippendorff's Alpha for ordinal data.
    """
    # Calculate pairwise differences
    n_items, n_raters = data.shape
    disagreements = 0
    total_pairs = 0

    for i in range(n_items):
        for j in range(n_raters):
            for k in range(j + 1, n_raters):
                disagreements += (data[i, j] - data[i, k]) ** 2
                total_pairs += 1

    observed_disagreement = disagreements / total_pairs

    # Compute expected disagreement 
    values = np.unique(data)
    value_counts = {value: np.sum(data == value) for value in values}
    total_values = np.sum(list(value_counts.values()))

    expected_disagreement = 0
    for value1 in values:
        for value2 in values:
            expected_disagreement += (value1 - value2) ** 2 * (value_counts[value1] / total_values) * (
                        value_counts[value2] / total_values)

    alpha = 1 - observed_disagreement / expected_disagreement
    return alpha


# Format data as numpy array and calculate Krippendorff's Alpha
ratings_array = iaa_full[['Rater 1', 'Rater 2']].to_numpy()
krippendorff_alpha = krippendorffs_alpha(ratings_array)

krippendorff_alpha
```
